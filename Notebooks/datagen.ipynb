```python
#importing the necessay dependancies
import random
random.seed(0)
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
import tensorflow as tf
tf.random.set_seed(0)
from tensorflow.keras import models, layers
from tensorflow.keras.preprocessing.image import  ImageDataGenerator
import splitfolders
from IPython.display import HTML
```

```python
#Defining the parameters
Image_size = 256
Channels = 3
Epochs = 15
```

```python
train_datagenerator = ImageDataGenerator(
    rescale=1.0/255,            # Normalize pixel values
    rotation_range=20,          # Rotate images up to 20 degrees
    width_shift_range=0.2,      # Horizontal shift
    height_shift_range=0.2,     # Vertical shift
    shear_range=0.2,            # Shear transformation
    zoom_range=0.2,             # Zoom in/out
    horizontal_flip=True        # Flip horizontally
)

train_generator = train_datagenerator.flow_from_directory(
    r"D:\Projects\Farm_intel\dataset\train",
    target_size=(Image_size, Image_size),
    batch_size=32,
    class_mode='sparse'
)
```

```python
test_datagenerator = ImageDataGenerator(
    rescale=1.0/255,            # Normalize pixel values
    rotation_range=20,          # Rotate images up to 20 degrees
    width_shift_range=0.2,      # Horizontal shift
    height_shift_range=0.2,     # Vertical shift
    shear_range=0.2,            # Shear transformation
    zoom_range=0.2,             # Zoom in/out
    horizontal_flip=True        # Flip horizontally
)

test_generator = test_datagenerator.flow_from_directory(
    r"D:\Projects\Farm_intel\dataset\test",
    target_size=(Image_size, Image_size),
    batch_size=32,
    class_mode='sparse'
)
```

```python
val_datagenerator = ImageDataGenerator(
    rescale=1.0/255,            # Normalize pixel values
    rotation_range=20,          # Rotate images up to 20 degrees
    width_shift_range=0.2,      # Horizontal shift
    height_shift_range=0.2,     # Vertical shift
    shear_range=0.2,            # Shear transformation
    zoom_range=0.2,             # Zoom in/out
    horizontal_flip=True        # Flip horizontally
)

val_generator = val_datagenerator.flow_from_directory(
    r"D:\Projects\Farm_intel\dataset\val",
    target_size=(Image_size, Image_size),
    batch_size=32,
    class_mode='sparse'
)
```
